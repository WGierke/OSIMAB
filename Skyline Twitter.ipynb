```python
import pandas as pd
import plotly.offline as offline
from plotly.graph_objs import Data, Figure
from tqdm import tqdm

from data_loading import get_df, get_labels_df
from detectors.skyline_detector import SkylineDetector
from plotting import get_data_scatter, get_anomaly_scatter, get_labels_scatter, createLayout, plot

data_df = get_df()
labels_df = get_labels_df(data_df)
scores_df = pd.DataFrame(columns=data_df.columns)
data_df = data_df[800:1500]

detector = SkylineDetector()

for index, row in tqdm(data_df.iterrows(), total=len(data_df)):
    scores_df = scores_df.append(detector.handle_row(row))
    
for value_key in data_df.columns:
    plot(data_df, scores_df, value_key, labels_df=labels_df, threshold=0.7)
```

```python
traces = []
for col in data_df.columns:
    traces.append(get_data_scatter(data_df, col))
data = Data(traces)
layout = createLayout(title=col)
fig = Figure(data=data, layout=layout)
offline.iplot(fig)
```
