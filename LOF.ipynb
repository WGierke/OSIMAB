```python
df = get_df('data/realTweets.csv')[:2000]
labels_df = get_labels_df(df)
scores_df = pd.DataFrame(columns=df.columns)
```

```python
detector = LofDetector(neighbors=200)
for index, row in tqdm(df.iterrows(), total=len(df)):
    scores_df = scores_df.append(detector.handle_row(row))
```

```python
plot(df, scores_df, 'value_GOOG', labels_df=labels_df, threshold=0.5)
print((scores_df["value_AAPL"] > 0).sum(), (scores_df["value_AAPL"] < 0).sum())
```
